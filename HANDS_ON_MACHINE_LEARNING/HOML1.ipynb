# Hands On Machine Learning with Scikit-Learn, Keras & Tensorflow

## 사이킷런, 케라스, 텐서플로2를 활용한 머신러닝 스터디 노트

머신러닝이 활용된 예제
* 광학 문자 판독기 (OCR)
* 스팸 메일, 문자 필터
* 콘텐츠 추천
* 음성 검색

### 머신러닝이란?

**명시적인 프로그래밍 없이 컴퓨터가 학습하는 능력을 갖추게 하는 것**

### 왜 머신러닝을 사용하는가?

전통적인 프로그래밍 기법으로 스팸 메일 필터를 만든다고 가정해보자.  
다음과 같은 4단계로 스팸 메일 필터를 만들 수 있다.

1. 스팸 메일에서 공통적으로 드러나는 특징들을 찾아낸다.

2. 해당 특징들을 처리하는 알고리즘을 처리한다.

3. 프로그램이 충분한 성능이 나타날 때까지 1, 2 단계를 반복한다.

당신은 성공적으로 프로그램을 런칭하였다.  
**하지만 이를 눈치챈 스팸 메일 전송자들이 단어의 패턴을 다르게 스팸 메일을 보내기 시작한다.**  

4. 스팸 메일 전송자와 영원한 싸움을 시작한다. (1, 2, 3 단계 반복)

반면, 머신러닝을 이용한 스팸 메일 필터는 스팸 메일에 자주 나타나는 특징을 감지하여 어떤 것이 좋은 판단기준인지 자동으로 학습합니다.
